# Notebook for Decision Tree model

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading the data to dataframe

```python
life_expectancy = pd.read_csv("./Life_Expectancy_Data.csv")
```

# Dropping rows with NaN values

```python
life_expectancy.replace(0,np.nan)
```

```python
life_expectancy = life_expectancy.dropna()
```

# Heatmap

```python
plt.figure(figsize = (20,20))
corr_matrix = life_expectancy.corr()
sns.heatmap(corr_matrix, annot = True)
plt.show()
```

```python
sns.scatterplot(data = life_expectancy, y=life_expectancy["Life expectancy"],x=life_expectancy["Income composition of resources"])
```

# Converting categorical data to independent variables

```python
life_expectancy = pd.get_dummies(life_expectancy, columns = ["Status"])
```

# Dividing the data into dependent and independent variables

```python
X = life_expectancy.drop(columns = ["Life expectancy"])
```

```python
y = life_expectancy[["Life expectancy"]]
```

```python
X
```

```python
y
```

# Converting data to numpy arrays

```python
X = np.array(X)
```

```python
y = np.array(y)
```

# Splitting the data into training and testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

# Scaling data

```python
from sklearn.preprocessing import StandardScaler

scaler_X = StandardScaler()
X_train = scaler_X.fit_transform(X_train)
X_test = scaler_X.transform(X_test)

scaler_y = StandardScaler()
y_train = scaler_y.fit_transform(y_train)
y_test = scaler_y.transform(y_test)
```

# Fitting the data to regression model

```python
from sklearn.tree import DecisionTreeRegressor

regression_model = DecisionTreeRegressor()
regression_model.fit(X_train, y_train)
```

```python
accuracy = regression_model.score(X_test, y_test)
```

```python
accuracy
```

```python
y_predict = regression_model.predict(X_test)
plt.plot(y_predict,y_test,"^",color = "r")
```
