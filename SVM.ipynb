```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
life_expectancy = pd.read_csv("./Life_Expectancy_Data.csv")
```

```python
life_expectancy.replace(0,np.nan)
```

```python
life_expectancy = pd.get_dummies(life_expectancy, columns = ["Status"])
```

```python
life_expectancy.apply(lambda x: x.fillna(x.mode()),axis = 0)
```

```python
life_expectancy["Adult Mortality"].mode()
```

```python
X = life_expectancy.drop(columns = ["Life expectancy"])
```

```python
y = life_expectancy[["Life expectancy"]]
```

```python
X = np.array(X)
```

```python
y = np.array(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler

scaler_X = StandardScaler()
X_train = scaler_X.fit_transform(X_train)
X_test = scaler_X.transform(X_test)

scaler_y = StandardScaler()
y_train = scaler_y.fit_transform(y_train)
y_test = scaler_y.transform(y_test)
```

```python
from sklearn.svm import SVR

regression_model = SVR(kernel="rbf",epsilon= 0.07)
regression_model.fit(X_train, y_train)
```

```python
regression_model.score(X_test, y_test)
```

```python
y_predict = regression_model.predict(X_test)
plt.plot(y_predict,y_test,"^",color = "r")
```
